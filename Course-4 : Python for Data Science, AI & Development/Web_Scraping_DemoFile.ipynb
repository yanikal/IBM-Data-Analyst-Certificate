```python
# Step 1: Importing libraries
from bs4 import BeautifulSoup #bs4 is a library, BeautifulSoup is a portion from bs4
import requests #request is a library, extract the content from the html of any website
```

```python
# Step 2: Adding URL
link = "https://ravindrababuravula.com/courses.php"
data = requests.get(link)
```

```python
# Step 3: Creating Soup object
soup = BeautifulSoup(data.text, "html.parser") #parser is a method to persue your data, there are 5 (google)
#soup.prettify #you can look at the html strucure/code without visiting the webpage
#soup.title
#print(soup.get_text())
```

```python
# Step 4: Retrieve the Data

#soup.find('a',class_="title").text
for i in soup.find_all('a', class_="title") : #a tag is for hyperlink
  print(i.text) 
  #print(i)
```

```python
for d in soup.find_all('div', class_="description"): #description for all the hyperlinks
  print(d.text) 
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from PIL import Image
from io import BytesIO
from IPython.display import display


url = 'https://ravindrababuravula.com/courses.php'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
img_tags = soup.find_all('img')


for _IMAGE in img_tags:
    imgURL = _IMAGE.get('src')
    imgURL = urljoin(url, imgURL)
#     Visulazie the img
    img_response = requests.get(imgURL)
    if img_response.status_code == 200:
            img = Image.open(BytesIO(img_response.content))
            display(img)
#             filename = img_url.split('/')[-1]
#             with open(filename, 'wb') as f:
#                 f.write(img_response.content)
```
